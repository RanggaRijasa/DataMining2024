```python
#g
fighters_df = mlbb_hero_df[mlbb_hero_df['role'] == 'fighter']
tanks_df = mlbb_hero_df[mlbb_hero_df['role'] == 'tank']

fighter_avg_atk = fighters_df['physical_atk'].mean()
tank_avg_atk = tanks_df['physical_atk'].mean()

print(f"Average physical attack for Fighters: {fighter_avg_atk}")
print(f"Average physical attack for Tanks: {tank_avg_atk}")

if fighter_avg_atk > tank_avg_atk:
    print("Fighters have higher average physical attack than Tanks.")
elif tank_avg_atk > fighter_avg_atk:
    print("Tanks have higher average physical attack than Fighters.")
else:
    print("Fighters and Tanks have the same average physical attack.")
```

7. I have a character with this specification:
defense_overall 6.0<br>
offense_overall 7.0<br>
skill_effect_overall 4.0<br>
difficulty_overall 3.0<br>
movement_spd 250<br>
magic_defense 9<br>
mana 380<br>
hp_regen 37<br>
physical_atk 119<br>
physical_defense 24<br>
hp 2689<br>
attack_speed 0.75<br>
mana_regen 16.0<br>

Please predict its role using DecisionTree classification algorithm <br>

```python
# 7

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Encode the 'role' column
encoder = LabelEncoder()
y = encoder.fit_transform(mlbb_hero_df['role'])

# Separate features and target variable
X = mlbb_hero_df[['defense_overall', 'offense_overall', 'skill_effect_overall', 'difficulty_overall', 
         'movement_spd', 'magic_defense', 'mana', 'hp_regen', 'physical_atk', 
         'physical_defense', 'hp', 'attack_speed', 'mana_regen']]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 10)

# Create a Decision Tree Classifier
model = DecisionTreeClassifier()

# Train the model
model.fit(X_train,y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Predict the role for the given character
character = [[6.0,7.0,4,3,250,9,380,37,119,24,2689,0.75,16]]
predicted_role = model.predict(character)

# Get the predicted role (dicoding)
predicted_role_name = encoder.inverse_transform(predicted_role)
print("Predicted Role:", predicted_role_name[0])
```
